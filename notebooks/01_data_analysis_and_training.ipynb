# Notebook 01: Phân tích dữ liệu & Huấn luyện (Data Analysis & Training)

Notebook này sử dụng `vtrace.data_utils` để phân tích cấu trúc dữ liệu của các scene, và `vtrace.trainer` để chạy pipeline huấn luyện 3D Gaussian Splatting.

```python
import os
import logging
from vtrace import data_utils, trainer

logging.basicConfig(level=logging.INFO)

base_data_dir = "../VAI_NVS_DATA/phase1/public_set"
scenes = data_utils.list_scenes(base_data_dir)
print(f"Found {len(scenes)} scenes: {scenes}")
```

### Phân tích một scene mẫu

```python
if scenes:
    sample_scene = scenes[0]
    scene_dir = os.path.join(base_data_dir, sample_scene)
    stats = data_utils.analyze_scene(scene_dir)
    print(f"Stats for {sample_scene}:")
    for k, v in stats.items():
        print(f"  {k}: {v}")
```

### Khởi chạy Huấn luyện (Training Pipeline)

Vòng lặp bên dưới sẽ huấn luyện mô hình cho từng scene. Bạn có thể giảm `iterations` xuống thấp (vd: 7000) để test nhanh, hoặc để mặc định 30000 cho chất lượng tốt nhất.

```python
import os
import sys

# Đảm bảo import được module từ thư mục src
sys.path.append(os.path.abspath(os.path.join("..")))
from src.vtrace import trainer

# Chạy huấn luyện (ví dụ: huấn luyện scene đầu tiên để test)
if scenes:
    sample_scene = scenes[0]
    scene_dir = os.path.join(base_data_dir, sample_scene)
    
    # Đặt output ra ngoài thư mục gốc
    output_dir = os.path.join("../output", sample_scene)
    
    print(f"Bắt đầu huấn luyện {sample_scene}...")
    
    # Gọi hàm train_scene với gs_path trỏ đúng về thư mục src/vtrace/...
    trainer.train_scene(
        scene_dir=scene_dir, 
        output_dir=output_dir, 
        iterations=7000, 
        resolution=1, 
        gs_path="../src/vtrace/gaussian-splatting"  # <--- Đã sửa dòng này
    )
    print("Huấn luyện hoàn tất!")
```
